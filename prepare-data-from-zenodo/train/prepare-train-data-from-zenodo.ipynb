# This notebook extracts train data at a given acquisition time from the Zenodo dataset and prepares it for the production pipeline.

```python
import numpy as np
import json
import h5py
```

```python
from ddae1d.paths import PROJECT_ROOT
```

```python
with open('config.json', 'r') as f:
    config = json.load(f)

acq_time = config["acq_time"]
n_reps = config["n_reps"]
```

```python
train_data_path = PROJECT_ROOT / f"data/from-zenodo/train-data-{acq_time}ms.h5"
with h5py.File(train_data_path, 'r') as h5file:
    raman_shift = h5file['raman_shift'][:]
    train_data = h5file['raw_spectra'][:]
```

Keeping only the last `n_reps` repetitions for data to create the trainset for Noise2Noise training.

```python
trainset = train_data[: , -n_reps:, :]
```

```python
np.save(PROJECT_ROOT / "data" / "raw" / "trainset"/ "raw-trainset.npy", trainset)
```
